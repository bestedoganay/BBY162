```
from google.colab import drive
drive.mount('/gdrive')
dosya = "/gdrive/My Drive/Colab Notebooks/finalprojesi/veritabani.txt"

print ("Kütüphane kataloğuna hoşgeldiniz.")

class katalog:

     def __init__(self):
        kitaplar = []
        kitaplik = open(dosya,"r",encoding = "UTF-8")
        for i in kitaplik.read().strip().split("\n"):
            kitap = i.strip().split("|")
            kitaplar += [kitap]
        kitaplik.close()
        self.kitaplar = kitaplar
        kitaplar = None


     def yenikayit(self):
        
        kitapAdi = input("Eklemek istediğiniz kitabın adını giriniz: ")
        yazarAdi = input("Eklemek istediğiniz kitabın yazarını giriniz: ")
        yayinyili = input("Eklemek istediğiniz kitabın yayın tarihini giriniz: ")
        yayinevi = input("Eklemek istediğiniz kitabın yayınevini giriniz: ")
        ISBN = input("Eklemek istediğiniz kitabın ISBN numarasını giriniz:")
        kitapekleme = (kitapAdi+ "|"+ yazarAdi + "|" + yayinyili + "|" + yayinevi +"|" +ISBN+ "\n")
        print("Kayıt Kataloğa Eklendi")
        kitaplik = open(dosya,"a",encoding = "UTF-8")
        kitaplik.write( "\n"+ kitapekleme)
        kitaplik.close()
        

     def duzenle(self):
        kitaplik=open(dosya, "r")
        print(kitaplik.read())
        kitaplik.close() 
        line= int(input("Değiştirmek istediğiniz satırı yazınız: "))
        kitaplik1=open(dosya, "r")
        liste=kitaplik1.readlines()
        kitaplik1.close()
        del liste[line]
        kitapAdi=input("Kitap Adını Giriniz: ")
        yazarAdi=input("Yazar Adını Giriniz: ")
        yayinyili=input("Yayın Tarihini Giriniz: ")
        yayinevi=input("Yayınevini Giriniz: ") 
        ISBN = input ("ISBN Giriniz:") 
        liste.insert(line, kitapAdi+ "|"+yazarAdi+ "|"+yayinyili+ "|"+yayınevi+"|" +ISBN+  "\n")
        print("Kayıt Başarıyla Düzenlendi")
        kitaplik2=open(dosya, "w")
        kitaplik2.writelines(liste)
        kitaplik2.close()
   
        
     def kayitsil(self):
        silinecek = input("Silmek İstediğiniz Kitabın Bilgilerini Giriniz: ")
        if len(self.kitaplar) < 1:
            return None

        for kitap in self.kitaplar:
            if silinecek in kitap:
                self.kitaplar.remove(kitap)
                break

        kitaplik = open(dosya,"a",encoding = "UTF-8")
        for kitap in self.kitaplar:
            kitaplik.write(f'{kitap[0]}|{kitap[1]}|{kitap[2]}|{kitap[3]|{kitap[4]}}')
            if not kitap == self.kitaplar[-1]:
                kitaplik.write("\n")
                print("Kitap Başarıyla Silindi!")
        kitaplik.close()

     def kitaplistele(self):
        if len(self.kitaplar) < 1:
            print("Kitap kaydı bulunmamaktadır. ")
            return None

        bos  = ""
        for kitap in self.kitaplar:
            bos += ('kitapAdi:' +kitap[0]+ '\n')
            bos += ('yazarAdi:' +kitap[1]+'\n')
            bos += ('yayinyili:' +kitap[2]+ '\n')
            bos += ('yayinevi:' +kitap[3]+ '\n')
            bos += ('ISBN:' +kitap[4]+ '\n')
            bos += ("-"*60 + "\n")
        print(bos)
     
     def kitapara(self):
        aranacak = input("Aranacak kelime giriniz:")
        bos = ""

        for kitap in self.kitaplar:
            if len(self.kitaplar) > 0 and aranacak in kitap:
                bos += ('kitapAdi:' +kitap[0]+ '\n')
                bos += ('yazarAdi:' +kitap[1]+ '\n')
                bos += ('yayinyili:' +kitap[2]+ '\n')
                bos += ('yayinevi:' +kitap[3]+ '\n')
                bos += ('ISBN:' +kitap[4]+ '\n')
                print(bos)
        if bos== "":
            print("Aradığınız Kitap Katalogda Bulunamadı.")

     def sonuc(self):
        while True:

             choose =input(" Yeni kitap eklemek İçin: 1, \n Kitapları Listelemek İçin: 2,\n Kitap Aramak İçin: 3,\n Kitap Düzenlemek İçin: 4,\n Kitap silmek için: 5, \n çıkıs 0 \n")

             if choose == "1":
                self.yenikayit()
             elif choose == "2":
                self.kitaplistele()
             elif choose == "3":
                self.kayitsil()
             elif choose == "4":
                self.duzenle()
             elif choose == "5":
                self.kayitara() 
             elif choose == "0":
                 
                break

    
   
katalog().sonuc()
```
